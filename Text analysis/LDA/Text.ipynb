```python
trans_prob = np.zeros((16,16))
```

```python
for i in next_states:
    pos = next_states[i]
    for j in pos:
        trans_prob[i,j] = 1
```

```python
row_sums = trans_prob.sum(axis = 1)
```

```python
trans_prob = trans_prob / row_sums[:, np.newaxis]
```

```python
p = np.array([[0,4,1,0], [0,5,2,1], [1,6,3,2], [2,7,3,3], [4,8,5,0], [5,5,5,5], [5,10,7,2], [7,7,7,7], [8,12,9,4], 
              [8,13,10,5], [9,14,11,6], [11,11,11,11], [12,12,12,12], [12,13,14,9], [13,14,15,10], [5,7,11,12]])
```

```python
p
```

```python
Q = np.zeros((16,4))
```

```python
max_q = np.max(Q, axis = 1)
temp1 = np.zeros((16,4))
for i in xrange(len(max_q)):
    temp1 += (p==i) * max_q[i]
Q = r + gamma * temp1
```

```python
Q
```

```python
count = 0
while True:
    count += 1
    Q_old = Q[:]
    max_q = np.max(Q, axis = 1)
    temp1 = np.zeros((16,4))
    for i in xrange(len(max_q)):
        temp1 += (p == i) * max_q[i]
    Q = r + gamma * temp1
    if np.sum(abs(Q_old - Q)) < 1e-10:
        break
print count
```
